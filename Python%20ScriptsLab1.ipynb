```python
import numpy as np
import matplotlib.pyplot as plt
x_data = np.arange(-2, 2, 0.01)
y_data = np.array([])
y = np.array([])
for x in x_data:
    for i in range(100):
        y = np.append(y, np.cos(np.pi*x*3**i)*0.5**i)
    y_sum = np.sum(y)
    y_data = np.append(y_data, y_sum)
    y = np.array([])
plt.plot(x_data, y_data)
plt.grid(True)
plt.show()
```
